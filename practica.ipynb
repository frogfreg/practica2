```python
tweet_01_tokens = tokenizer.tokenize(tweets[0].lower())
print(tweet_01_tokens)
```

```python
tweet_01_tokens_wout_stopwords = []
for word in tweet_01_tokens:
    if word not in stopwords.words('spanish'): tweet_01_tokens_wout_stopwords.append(word)
print(tweet_01_tokens_wout_stopwords)        
```

```python
tweets_tokens = []
for tweet in tweets:
        tweets_tokens.append(tokenizer.tokenize(tweet.lower()))
print(tweets_tokens)
```

```python
tweets_tokens_wout_stopwords = []
for x in range (25):
    tweets_tokens_wout_stopwords.append([])
for i in range(25):
    for word in tweets_tokens[i]:
        if word not in stopwords.words('spanish'): tweets_tokens_wout_stopwords[i].append(word)
print(tweets_tokens_wout_stopwords)


        
```

```python
dicc_termns = {}

for tweet in tweets_tokens_wout_stopwords:
    for word in tweet:
        
#        print("EVALUAR:", word, "EN", text)
        
        if(word in dicc_termns):#incrementar palabras al diccionario
            dicc_termns[word] = dicc_termns[word] + 1
            
#            print(word, "IN", "dicc_termns")
            
        elif(word not in dicc_termns):#agregar palabras al diccionario        
            dicc_termns[word] = 1
            
#            print(word, "NOT IN", "dicc_termns")            

print(len(dicc_termns))
dicc_termns
```

```python
matrix = np.zeros((len(tweets_tokens_wout_stopwords), len(dicc_termns)))
```

```python
i = 0
j = 0

for word_termns in dicc_termns: #dicc_termns todos los términos
#    print()
    for tweet in tweets_tokens_wout_stopwords: #dicc_texts todos los textos
#        print("EVALUAR:", word_termns, "EN: ", word_texts)
        if(word_termns in tweet): #si está
            print(word_termns, "IN", f"tweet{tweets_tokens_wout_stopwords.index(tweet)}")
            
            matrix[j, i] = 1
            
        elif(word_termns not in tweet): # si no está
            print(word_termns, "NOT IN", f"tweet{tweets_tokens_wout_stopwords.index(tweet)}")
            
            matrix[j, i] = 0
            
            
        print("se agregó: ", matrix[j,i], "en: ", j, i)
            
        j = j + 1
        
    j = 0
    i = i + 1
```
